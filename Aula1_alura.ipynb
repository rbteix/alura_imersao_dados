## Aula 04 - Machine Learning

```
exams_inp = ["NU_NOTA_CH","NU_NOTA_CN", "NU_NOTA_REDACAO"]
exams_out = "NU_NOTA_MT"
data_nogrades_zero= data_nogrades_zero[exams].dropna()
grades_inp = data_nogrades_zero[exams_inp]
grades_out = data_nogrades_zero[exams_out]
```

```
grades_inp
```

```
x = grades_inp
y = grades_out
```

```
#Split in train and test
from sklearn.model_selection import  train_test_split
train_test_split(x, y)
```

```
seed = 234
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = seed)
```

```
x_train.head()
```

```
#Regression Model (svm)
#The model will predict the math grade based on the four grade subjects.
from sklearn.svm import LinearSVR
model = LinearSVR(random_state=seed)
model.fit(x_train, y_train)
```

```
math_predict = model.predict(x_test)
```

```
y_test
```

```
plt.figure(figsize=(8,8))
sns.scatterplot(x=math_predict, y= y_test)
plt.xlim((-50, 1050))
plt.ylim(( -50,1050))
```

```
# y= y_test - math_predict (error) by real grades. Tendencious model,
# the grater the grade  grater the prediction and the oposit applies as well.
plt.figure(figsize=(8,8))
sns.scatterplot(x=y_test, y= y_test - math_predict)
```

```
results = pd.DataFrame()
results["Real"] = y_test
results["Predict"] = math_predict
results["dif"] = results["Real"] - results["Predict"]
results["sqd_dif"] = (results["Real"] - results["Predict"])**2
```

```
#Decision Tree model - less dependent from randomness

from sklearn.tree import DecisionTreeRegressor
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
model_tree = DecisionTreeRegressor(max_depth = 3)
model_tree.fit(x_train, y_train)
math_predict_tree = model_tree.predict(x_test)
mean_squared_error(y_test, math_predict_tree)
```

```
#Model validation
from sklearn.model_selection import cross_validate

model_tree = DecisionTreeRegressor(max_depth=2)
# grade mathod, score parameter: give me the error, I give you the socre
result_cross = cross_validate(model_tree, x, y, cv = 10, scoring="neg_mean_squared_error") #the higher the better, implementation rule
mean1 = (result_cross["test_score"]* -1).mean() #the lower the better 
sd = (result_cross["test_score"]*-1).std()
inf_lim = mean1 - (2*sd)
upper_lim = mean1 + (2*sd)

print(f"Intervalo de confiança {upper_lim} - {inf_lim}") #If you execut 100 times in 95 the right answer must be there (see)
```

```
def calc_mse(result_cross):
  mean1 = (result_cross["test_score"]* -1).mean() 
  sd = (result_cross["test_score"]*-1).std()
  inf_lim = mean1 - (2*sd)
  upper_lim = mean1 + (2*sd)

  print(f"Confidence Interval {inf_lim} - {upper_lim}")
```

```
from sklearn.model_selection import cross_validate

model_tree = DecisionTreeRegressor(max_depth=2)
# grade mathod, score parameter: give me the error, I give you the socre
result_cross = cross_validate(model_tree, x, y, cv = 10, scoring="neg_mean_squared_error") #the higher the better, implementation rule
calc_mse(result_cross)
```
